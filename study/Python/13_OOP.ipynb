### Object Oriented Programming:

```python
class Smartphone:
    def __init__(self, name, ram, storage, camera): # constructor method 
        self.name = name # attributes 
        self.ram = ram 
        self.storage = storage
        self.camera = camera
    
    def call(self, phone_num): # Methods
        print(f'calling to {phone_num}...')

    def message(self, phone_num, message): # Methods
        print(f'message: {message} sent to phone number: {phone_num}')
```

```python
mi = Smartphone(name="MI", ram=4, storage=32, camera=12)
```

```python
mi.call(phone_num=39248739)
```

```python
mi.message(phone_num=239843, message='Thank you for Being my 100th subscriber.')
```

### Multiple inheritance

```python
class Animal: # base/parent 
    def __init__(self):
        self.num_of_legs = 4
    def speak(self):
        return 'Animal Sound'

class Alive:
    def __init__(self):
        self.isalive = True 

class Dog(Animal, Alive): # subclass / child class 
    def __init__(self):
        Alive.__init__(self)
        Animal.__init__(self)
    def speak(self): # method overriding
        return 'Woof'   
```

```python
d = Dog()
```

```python
d.num_of_legs
```

```python
d.isalive
```

```python
d.speak()
```

### Multi-level inheritance

```python
class Animal: # base/parent 
    def __init__(self):
        self.num_of_legs = 4
    def speak(self):
        return 'Animal Sound'

class Alive(Animal):
    def __init__(self):
        super().__init__()
        self.isalive = True 

class Dog(Alive): # subclass / child class 
    def __init__(self):
        super().__init__()
    def speak(self): # method overriding
        return 'Woof'  
```

```python
d2 = Dog()
```

```python
d2.num_of_legs
```

```python
d2.isalive
```

```python

class Calculator:
    def add(self, x, y):
        return x + y

    def add(self, x, y, z):
        return x + y + z

calc = Calculator()

print(calc.add(2, 3))  
print(calc.add(2, 3, 4)) 
    
```

```python
class Calculator:
    def add(self, x, y, z=None):
        if z is None:
            return x + y
        else:
            return x + y + z
```
